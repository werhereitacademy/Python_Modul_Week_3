## Python_Modul_Hafta_3


Soru 1: Görev Yöneticisi Uygulamasi
Proje Açiklamasi:
Bu ödevde, Python programlama dili kullanarak bir görev yöneticisi uygulamasi oluşturacaksiniz. Bu uygulama, kullanicilarin görev eklemesine, tamamlamasina, silmesine ve görevlerini listelemesine olanak taniyacak.

Gereksinimler:

1- Görevler bir Python listesinde saklanacak ve her görev bir sözlük (dictionary) olarak temsil edilecek. Her görev şu özelliklere sahip olmalidir:

Sira Numarasi (Otomatik olarak atanacak)

Görev Adi

Durum (Tamamlandi, Bekliyor veya Silindi)

2- Kullanicinin gerçekleştirebileceği işlemler:

Yeni bir görev eklemek

Bir görevi tamamlamak

Bir görevi silmek

Tamamlanan görevleri listelemek

Tüm görevleri durumlariyla birlikte listelemek

Çikiş yapmak

3- Görevlere, eklendikleri siraya göre otomatik olarak bir sira numarasi verilmelidir.

4- Yeni eklenen görevler, silinmiş görevlerin numaralarinin yerine kaydedilmelidir.

5- Görevler listelenirken sira numarasina göre sirali olarak gösterilmelidir.

6- Her işlemden sonra kullaniciya uygun bir geri bildirim verilmelidir. Örneğin, yeni bir görev eklendiğinde, görev eklendiğini belirten bir mesaj gösterilmelidir.

```python
gorevler = []
bos_siralar = []  
sira_numarasi = 0  

def gorev_ekleme():
    global sira_numarasi

    gorev_adi = input("Görev adi giriniz: ")
    gorev_durumu = "bekliyor"  

    if bos_siralar:
        yeni_sira = bos_siralar.pop(0)
        gorevler[yeni_sira] = {
            "numara": yeni_sira + 1,
            "ad": gorev_adi,
            "durum": gorev_durumu
        }
    else:
        gorevler.append({
            "numara": sira_numarasi + 1,
            "ad": gorev_adi,
            "durum": gorev_durumu
        })
        sira_numarasi += 1

    print(f"Görev eklendi: {gorev_adi}")

def gorev_tamamlama():
    try:
        sira = int(input("Tamamlanacak görevin numarasini giriniz: "))
        for g in gorevler:
            if g and g["numara"] == sira:
                g["durum"] = "tamamlandi"
                print(f"{sira} numarali görev tamamlandi.")
                return
        print("Görev bulunamadi.")
    except ValueError:
        print("Geçerli bir sayi girin.")

def gorev_silme():
    try:
        numara = int(input("Silinecek görev numarasini girin: "))
        for i, g in enumerate(gorevler):
            if g and g["numara"] == numara:
                g["durum"] = "Silindi"
                bos_siralar.append(i)
                print(f"{numara} numarali görev silindi.")
                return
        print("Görev bulunamadi.")
    except ValueError:
        print("Geçerli bir sayi girin.")

def tamamlanan_gorevleri_listeleme():
    tamamlanmislar = [g for g in gorevler if g and g["durum"] == "tamamlandi"]
    if not tamamlanmislar:
        print("Tamamlanmiş görev yok.")
    else:
        print("Tamamlanmiş Görevler:")
        for g in tamamlanmislar:
            print(f'{g["numara"]}. {g["ad"]} - {g["durum"]}')

def tum_gorevleri_listeleme():
    aktif_gorevler = [g for g in gorevler if g and g["durum"] != "Silindi"]
    if not aktif_gorevler:
        print("Görev bulunamadi.")
    else:
        print("Tüm Görevler:")
        for g in aktif_gorevler:
            print(f'{g["numara"]}. {g["ad"]} - {g["durum"]}')


# Ana döngü
while True:
    try:
        kullanici_girisi = int(input("""
            Yapmak istediğiniz işlemin numarasini giriniz: 
            1 - Görev ekleme 
            2 - Görev tamamlama
            3 - Görev silme 
            4 - Tamamlanan görevleri listeleme 
            5 - Tüm görevleri listeleme                      
            6 - Çikiş yapma
            >>> """))

        if kullanici_girisi == 1:
            gorev_ekleme()
        elif kullanici_girisi == 2:
            gorev_tamamlama()
        elif kullanici_girisi == 3:
            gorev_silme()
        elif kullanici_girisi == 4:
            tamamlanan_gorevleri_listeleme()
        elif kullanici_girisi == 5:
            tum_gorevleri_listeleme()
        elif kullanici_girisi == 6:
            print("Çikiliyor...")
            break
        else:
            print("Geçersiz seçim. Lütfen 1-6 arasinda bir sayi girin.")
    except ValueError:
        print("Lütfen bir sayi girin.")
```

```python
#question 8
def word_frequency(text):
    words = text.split()
    frequency = {}
    for word in words:
        if word in frequency:
            frequency[word] += 1 
        else:
            frequency[word] = 1 
        
    return frequency
text = "this is a test this is only a test"
print(word_frequency(text))

```
