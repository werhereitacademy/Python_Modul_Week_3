```python
##programe task manager application : this program allow user to add,delete,complete and list thier view. 
##auther :Mokhtar
TaskList = [] # list to store task
TaskCount = 0 # variable to store task count
#task_number_delete = None # variable to store task number to delete
# function to add task
def add_task(task_name):
    global TaskCount
    ### Punt num 4 : check if task number is delete or not :=========================
    ###=========================================================
    used_task_numbers = [task['Task Number'] for task in TaskList] # get list of used task numbers
    task_number_delete = None # initialize task number to delete
    for task1 in range(1,TaskCount+1): # check if task number not in task list
        if task1 not in used_task_numbers:
            task_number_delete = task1 # 
            break
    if task_number_delete is not None:
        task = {
            'Task Number': task_number_delete,
            'Task Name': task_name,
            'Status': 'Pending' 
        }
        TaskList.append(task) # add task to list
        print(f" ✅ Task '{task_name}' added with Task Number {task_number_delete}.") # print message to user
    else:
        TaskCount += 1
        task ={
            'Task Number': TaskCount,
            'Task Name': task_name,
            'Status': 'Pending'         
        }
        TaskList.append(task) # add task to list
        print(f" ✅ Task '{task_name}' added with Task Number {TaskCount}.") # print message to user
   
##=========================================================
def list_tasks():
    if not TaskList:
        print("No tasks found!")
        return
    
    # Print table header
    print("\nTask List:")
    print("\n{:<10} {:<30} {:<15}".format("Number", "Task Name", "Status"))
    print("-" * 55)
    
    # Print each task
    for task in TaskList:
        print("{:<10} {:<30} {:<15}".format(
            task["Task Number"],
            task["Task Name"][:28] + ('..' if len(task["Task Name"]) > 28 else ''),
            task["Status"][:13] + ('..' if len(task["Status"]) > 13 else '')
        ))
    
    print(f"\nTotal Tasks: {len(TaskList)}")
## function to complete task
def compelete_task(task_number):
    for task in TaskList:
        if task['Task Number'] == task_number:
            task['Status'] = 'Completed'
            print(f"Task {task_number} marked as completed.")
            return
    print(f" ✅ Task {task_number} not found.")
##========================================================

## fuction to list completed tasks
def list_completed_tasks():
    completed_tasks = [task for task in TaskList if task['Status'] == 'Completed']
    if not completed_tasks:
        print("No completed tasks found.")
        return
    
    print("\nCompleted Tasks:")
    print("{:<10} {:<30} {:<15}".format("Number", "Task Name", "Status"))
    print("-" * 55)
    
    for task in completed_tasks:
        print("{:<10} {:<30} {:<15}".format(
            task["Task Number"],
            task["Task Name"][:28] + ('..' if len(task["Task Name"]) > 28 else ''),
            task["Status"][:13] + ('..' if len(task["Status"]) > 13 else '')
        ))
    print(f"\nTotal Completed Tasks: {len(completed_tasks)}")
    ##=======================================================

def delete_task1(task_number):
    for task in TaskList:
       # if task['Task Number'] == task_number:
        if task_number <1 or task_number > len(TaskList):
            print(f" ❌ Task {task_number} not found.")
            return
        elif task['Task Number'] == task_number:
         #   TaskList.remove(task)
            confirm = input(f"Are you sure you want to delete Task {task_number}: '{task['Task Name']}'? (y/n): ")
            if confirm.lower() == 'y':
              TaskList.remove(task)
              print(f" ✅ {task} :  deleted successfully!")
              return
            else:
              print(" ❌ Deletion Canceled.")
              return
    
##========================================================
## function to delete task.
'''
def delete_task(task_number):
    global TaskList
    # Check if task number exists in TaskList
    if task_numer <1 or task_number>len(TaskList):
    if task_number <1 or task_number > len(TaskList):
        print(f" ❌ Task {task_number} not found.")
        print('-----------')
        return
    else:
        print(f"Are you sure you want to delete Task {task_number}? (y/n): ")
        confirm = input()
        if confirm.lower() != 'y':
            print(" ❌ Task not deleted.")
            return
        else:
            TaskList = [task for task in TaskList if task['Task Number'] != task_number] # filter out the task to be deleted
    print(f"✅ Task {task_number} deleted successfully.")''' # print message to user
    # Delete the task from TaskList 
'''
    try:
        if task_number in TaskList:
                confirm = input(f"Are you sure you want to delete '{task_number}'? (y/n): ")
                if confirm.lower() == 'y':
                   TaskList.remove(task_number) # remove task from list
                   print(f"✅ {task_number} :  deleted successfully!")
                   #view_customer()##calling to show details.
                else:
                   print(" ❌ Task not found.")
    except ValueError:
            print("invalid inpt")'''
##========================================================
def list_tasks1():
    if not TaskList: # check if task list is empty
        print("No tasks available.")
        return
    print("\nTask List:")
    for task in TaskList: # iterate through task list
       # print(f"Task Number: {task['Task Number']}, Task Name: {task['Task Name']}, Status: {task['Status']}") # print task details
       
        print("{:<10} {:<30} {:<15}".format("Task Number", "Task Name", "Status"))
        print("-" * 50)

        print("{:<10} {:<30} {:<15}".format(
            task['Task Number'] ,
           task['Task Name'][:28] + ('..' if len(TaskList['Task Name']) > 28 else ''),
           TaskList['Status'][:13] + ('..' if len(TaskList['Status']) > 13 else '')))
        print(f"\nTotal Tasks: {len(TaskList)}")

#function main to display menu and get user input
##==================================================================
##=================== Main Function  ===========================
##===========================================================

def main():
    while True:
        print("\n Task Manger Aplication")
        print("1 : Add Task")
        print("2: complete Task")
        print("3 : Delete Task")
        print("4 : list Task ")
        print("5 : list Completed Task")
        print("6 : Exit")
        choice =input("Enter your choice (1-6):")
        if choice == '1':
            task_name = input("Enter task name")
            add_task(task_name)
        elif choice =='2':
            task_number=int(input("Enter task number to complete:"))
            compelete_task(task_number)
        elif choice == '3':
            task_number = int(input("Enter task number to delete:"))
            delete_task1(task_number)
        elif choice == '4':     
            list_tasks()
        elif choice == '5':
            list_completed_tasks()
        elif choice == '6':
            print("Exiting the program.")
            break
        else:
            print("Invalid choice. Please try again.")  
if __name__ =="__main__":
    main() # call main function to start the program # function to complete task
```
